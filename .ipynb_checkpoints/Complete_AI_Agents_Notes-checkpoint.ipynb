# ✅ **AI Agent**

An **AI Agent** is like a smart assistant that can **understand tasks, make decisions**, and **use tools to act** in the real world (e.g., search the web, send emails, analyze data).

---

# ✅ **Tools**

**Tools** are extra functions that an AI agent can use to perform tasks it can't do on its own.

- **Built-in Tools**: Already provided (e.g., calculator, browser, Python).
- **Custom Tools**: Created by you for special tasks (e.g., query a private database, call an API).

You can create custom tools in 3 ways:

1. `@tool` decorator  
2. `StructuredTool` + `Pydantic`  
3. Inheriting from `BaseTool` class  

---

# ✅ **Tool Binding**

**Tool Binding** means **connecting the tool to the agent**, so the agent knows the tool exists and can use it when needed.

---

# ✅ **Tool Calling**

**Tool Calling** happens when the **agent decides to use a tool** for a specific user query.  
**Example:** You ask, “What’s the weather in Paris?” → The agent calls the **weather tool**.

---

# ✅ **Tool Execution**

**Tool Execution** is when the **tool actually runs** the function and gives a result.  
**Continuing the example:** The **weather tool runs**, fetches Paris weather, and returns it to the agent.

---

# ✅ **AI Agent Flow**

---

# 1. Agent Executor
- Entry point for the agent system.

Next: **Receive User Query**

---

# 2. Receive User Query
- Gets the user input (e.g., "What's the population of Paris?").

Next: **Pass query + scratchpad to agent**

---

# 3. Pass Query + Scratchpad
- Sends user input + memory to the agent.

Next: **Agent Response**

---

# 4. Agent Response
- Agent thinks and responds with:
  - **Action** (if a tool is needed), or  
  - **Finish** (if final answer is ready)

---

# If Agent Action:
1. **Agent Action** – Decides tool + input  
2. **Execute Tool** – Runs the tool  
3. **Collect Observation** – Gets result  
4. **Update Scratchpad** – Saves result  

Goes back to **Agent Response**

---

# If Agent Finish:
1. **Agent Finish** – Forms final answer  
2. **Return Final Output** – Sends to user

---

# Overall Flow:

User → AgentExecutor → UserQuery → Agent →
[Action → Tool → Observe → Loop]

OR

[Finish → Output]
